```python
# %%
import numpy as np
import matplotlib.pyplot as plt
import os

# Load crash data
prefix = "instability_step_1839"
folder = "logs"

def load_csv(name):
    return np.loadtxt(os.path.join(folder, f"{prefix}_{name}.csv"), delimiter=",")

S = load_csv("S")
t = load_csv("t")
sigma = load_csv("sigma")
V_crash = load_csv("solution_row")
a = load_csv("a")
b = load_csv("b")
c = load_csv("c")
rhs = load_csv("rhs")
v_crash = load_csv("solution_row")



# check length
print("rhs shape:", rhs.shape)

print("S shape:", S.shape)
print("V_crash shape:", v_crash.shape)
print("S min/max:", np.min(S), np.max(S))
print("V_crash min/max:", np.min(v_crash), np.max(v_crash))
print("Any NaN in V_crash?", np.any(np.isnan(v_crash)))
print("Any inf in V_crash?", np.any(np.isinf(v_crash)))


# %%
# Plot solution at crash
plt.figure(figsize=(10, 5))
plt.plot(S, V_crash, label='Option Value')
plt.xlabel('Stock Price S')
plt.ylabel('Option Value')
plt.title('Option Value at Instability Step')
plt.grid(True)
plt.legend()
#plt.tight_layout()
plt.show()

# %%
# Plot volatility function
plt.figure(figsize=(10, 5))
plt.plot(S, sigma, label='Local Volatility', color='orange')
plt.xlabel('Stock Price S')
plt.ylabel('Volatility σ(S, τ)')
plt.title('Local Volatility at Instability Step')
plt.grid(True)
plt.legend()
#plt.tight_layout()
plt.show()

# %%
# Plot matrix coefficients A, B, C
plt.figure(figsize=(12, 6))
plt.plot(a, label='A (lower diag)', linestyle='--')
plt.plot(b, label='B (main diag)')
plt.plot(c, label='C (upper diag)', linestyle='-.')
plt.xlabel('Index i')
plt.ylabel('Coefficient Value')
plt.title('Tridiagonal Matrix Coefficients at Instability Step')
plt.grid(True)
plt.legend()
#plt.tight_layout()
plt.show()

# %%
# Plot RHS
plt.figure(figsize=(10, 5))
plt.plot(rhs, label='RHS Vector', color='purple')
plt.xlabel('Index i')
plt.ylabel('Value')
plt.title('Right-Hand Side Vector at Instability Step')
plt.grid(True)
plt.legend()
#plt.tight_layout()
plt.show()
```
